```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
movies=pd.read_csv("/content/drive/MyDrive/datasets/tmdb_5000_movies.csv")
credits=pd.read_csv("/content/drive/MyDrive/datasets/tmdb_5000_credits.csv")
```

```python
credits
```

```python
movies=movies.merge(credits,left_on='id',right_on='movie_id')
movies.head()
```

```python
movies=movies[['movie_id','genres','overview','title_x','keywords','cast','crew']]
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.shape
```

```python
movies.duplicated().sum()
```

```python
movies.rename(columns={'title_x':'title'},inplace=True)
```

```python
movies.sample()
```

```python
movies.iloc[280].genres
```

```python
import ast
def convert(obj):
  obj=ast.literal_eval(obj)
  l=[]
  for i in obj:
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  obj=ast.literal_eval(obj)
  l=[]
  counter=0
  for i in obj:
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['cast']
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
  obj=ast.literal_eval(obj)
  l=[]
  for i in obj:
    if i['job']== "Director":
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies['crew']
```

```python
movies.sample(2)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))

```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(stop_words="english",max_features=5000)
vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def steming(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return "  ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(steming)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),key=(lambda x:x[1]),reverse=True)
```

```python
def recommend(movie):
  index=new_df[new_df['title']==movie].index[0]
  distance=sorted(list(enumerate(similarity[index])),key=(lambda x:x[1]),reverse=True)
  for i in distance[1:10]:
    print(new_df.iloc[i[0]].title)
```

```python
recommend("The Dark Knight Rises")
```

```python
%%writefile app.py
import streamlit as st
import pickle
import requests

movies=pickle.load(open("movie_list.pkl","rb"))
similarity=pickle.load(open("similarity.pkl","rb"))


def fetch_poster(movie_id):
  url='https://api.themoviedb.org/3/movie/{}?api_key=2736a08daef7a534d3cf2d8c371e0427&language=en-US'.format(movie_id)
  response=requests.get(url)
  data=response.json()
  posterpath=data['poster_path']
  full_path="https://image.tmdb.org/t/p/w500/"+posterpath
  return full_path

def recommend(movie):
  index=movies[movies['title']==movie].index[0]
  movie_name_list=[]
  recommended_movie_poster=[]
  distance=sorted(list(enumerate(similarity[index])),key=(lambda x:x[1]),reverse=True)
  for i in distance[0:10]:
   movie_id=movies.iloc[i[0]].movie_id
   recommended_movie_poster.append(fetch_poster(movie_id))
   movie_name_list.append(movies.iloc[i[0]].title)
  return movie_name_list,recommended_movie_poster


st.header("movie recommendation system")

movie_lst=movies['title'].values
select_option=st.selectbox("select or type from the list",movie_lst)


if st.button("recommend"):
 movie_name_list,recommended_movie_poster=recommend(select_option)
 st.subheader("recommended movies")
 c1,c2,c3,c4,c5,c6,c7,c8,c9,c10=st.tabs(["Movie 1","Movie 2","Movie 3","Movie 4","Movie 5","Movie 6","Movie 7","Movie 8","Movie 9","Movie 10"])
 with c1:
  st.text(movie_name_list[0])
  st.image(recommended_movie_poster[0])
 with c2:
  st.text(movie_name_list[1])
  st.image(recommended_movie_poster[1])

 with c3:
  st.text(movie_name_list[2])
  st.image(recommended_movie_poster[2])

 with c4:
  st.text(movie_name_list[3])
  st.image(recommended_movie_poster[3])

 with c5:
  st.text(movie_name_list[4])
  st.image(recommended_movie_poster[4])

 with c6:
  st.text(movie_name_list[5])
  st.image(recommended_movie_poster[5])

 with c7:
  st.text(movie_name_list[6])
  st.image(recommended_movie_poster[6])
 with c8:
  st.text(movie_name_list[7])
  st.image(recommended_movie_poster[7])

 with c9:
  st.text(movie_name_list[8])
  st.image(recommended_movie_poster[8])

 with c10:
  st.text(movie_name_list[9])
  st.image(recommended_movie_poster[9])


```
